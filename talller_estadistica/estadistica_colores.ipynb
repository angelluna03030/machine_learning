```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_excel('impar.xlsx', sheet_name="Colores") 
```

```python

frecuencia = dataset['Coloress'].value_counts().reset_index()

frecuencia
```

```python

frecuencia.columns=['Colores', 'Frecuencia_absoluta']


frecuencia['Frecuencia_acomulada_absoluta'] = frecuencia['Frecuencia_absoluta'].cumsum()


suma_acumulada = frecuencia['Frecuencia_absoluta'].sum()

frecuencia['Frecuencia_relativa'] = frecuencia['Frecuencia_absoluta']/suma_acumulada

frecuencia['Frecuencia_acomulada_relativa'] = frecuencia['Frecuencia_relativa'].cumsum()

frecuencia
```

```python
fig, ax = plt.subplots()
plt.title('Colores vs Frecuencia absoluta')
plt.xlabel('Colores')
plt.ylabel('Frecuencia_absoluta')
ax.bar(frecuencia['Colores'].astype(str), frecuencia['Frecuencia_absoluta'], color='green') #Cambiar tipo de grafica
plt.xticks(rotation=45)
plt.show()
```

```python
fig, ax = plt.subplots()
plt.title('Colores vs Frecuencia relativa')
color = frecuencia['Colores'].astype(str)
frecuencia_relativa = frecuencia['Frecuencia_relativa']
ax.pie(frecuencia_relativa, labels=color,  autopct='%1.1f%%')
plt.show()
```

```python
moda= dataset ['Coloress'].mode()[0]
print("La moda de los colores es: ",moda)
```

En conclusin al graficar la frecuencia absoluta y relativa de los colores, se puede visualizar la distribucin de manera clara y comprensible. La frecuencia absoluta nos muestra cuntas veces aparece cada color en el conjunto de datos, mientras que la frecuencia relativa nos proporciona una perspectiva de la distribucin de estos colores en trminos de porcentajes del total. Para finalizar en cuanto a la moda encontramos que el color ms repetido es el Amarillo.
