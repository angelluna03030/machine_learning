```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_excel('impar.xlsx', sheet_name="Altura") 
```

```python

frecuencia = dataset['Alturas'].value_counts().reset_index()

frecuencia
```

```python

frecuencia.columns=['Alturas', 'Frecuencia_absoluta']


frecuencia['Frecuencia_acomulada_absoluta'] = frecuencia['Frecuencia_absoluta'].cumsum()


suma_acumulada = frecuencia['Frecuencia_absoluta'].sum()

frecuencia['Frecuencia_relativa'] = frecuencia['Frecuencia_absoluta']/suma_acumulada

frecuencia['Frecuencia_acomulada_relativa'] = frecuencia['Frecuencia_relativa'].cumsum()

frecuencia
```

```python
fig, ax = plt.subplots()
plt.title('Alturas vs Frecuencia absoluta')
plt.xlabel('Alturas')
plt.ylabel('Frecuencia_absoluta')
ax.bar(frecuencia['Alturas'].astype(str), frecuencia['Frecuencia_absoluta'], color='green') #Cambiar tipo de grafica
plt.xticks(rotation=45)
plt.show()
```

```python
fig, ax = plt.subplots()
plt.title('Alturas vs Frecuencia relativa')
plt.xlabel('Frecuencia relativa')
plt.ylabel('Alturas')
ax.barh(frecuencia['Alturas'].astype(str), frecuencia['Frecuencia_relativa'], color='blue')
plt.xticks(rotation=45)
plt.grid(True, axis='x')
plt.show()
```

```python
media = dataset['Alturas'].mean()
mediana = dataset['Alturas'].median()
minimo = dataset['Alturas'].min()
maximo = dataset['Alturas'].max()
moda = dataset['Alturas'].mode()[0]
varianza = dataset['Alturas'].var()

print("La media de las alturas es:  : ", media)
print("La mediana de las alturas es  : ", mediana)
print("La altura minima es  : ", minimo)
print("La altura maxima es  : ", maximo)
print("La moda es  : ", moda)
print("La varianza es : ", varianza)
```

En conclusin al graficar la frecuencia absoluta y relativa de las alturas, se puede visualizar la distribucin de manera clara y comprensible. La frecuencia absoluta nos muestra cuntas veces aparece cada altura en el conjunto de datos, mientras que la frecuencia relativa nos proporciona una perspectiva de la distribucin de estas alturas en trminos de porcentajes del total. Adems al calcular las estadsticas descriptivas, se puede evidenciar que la media de las alturas es de aproximadamente 1.7061818181818182, la mediana de las alturas es de aproximdamente  1.69, la altura mnima registrada es de 1.5, mientras que la altura mxima es de 1.93.
