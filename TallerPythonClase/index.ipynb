```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
CD = pd.read_excel('Datosestadisticos.xlsx', sheet_name='Cuantitativos_Discretos')
```

```python
frecuenciaCD =  CD['Talla_Zapatos'].value_counts().reset_index()
frecuenciaCD
```

```python
frecuenciaCD.columns=['Talla_zapatos', 'Frecuencia_absoluta']
frecuenciaCD['Frecuencia_acumulada_absoluta'] = frecuenciaCD['Frecuencia_absoluta'].cumsum()
```

```python
suma_acumulada = frecuenciaCD['Frecuencia_absoluta'].sum()
frecuenciaCD['Frecuencia_relativa'] = frecuenciaCD['Frecuencia_absoluta'] / suma_acumulada
```

```python
frecuenciaCD['Frecuencia_acumulada_relativa'] = frecuenciaCD['Frecuencia_relativa'].cumsum()
```

```python
plt.figure(figsize=(10, 5))
plt.bar(frecuenciaCD['Talla_zapatos'], frecuenciaCD['Frecuencia_absoluta'], color='skyblue')
plt.xlabel('Categoría')
plt.ylabel('Frecuencia Relativa')
plt.title('Gráfico de Barras - Frecuencia Absoluta')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(8, 8)) 
plt.pie(frecuenciaCD['Frecuencia_relativa'], labels=frecuenciaCD['Talla_zapatos'], autopct='%1.1f%%', startangle=140)
plt.title('Diagrama de Pastel - Frecuencia Relativa')
plt.axis('equal')
plt.show()
```

```python
datasetCC = pd.read_excel('Datosestadisticos.xlsx', sheet_name='Cuantitativos_Continuos')
```

```python
num_intervalos = int(1 + np.log2(len(datasetCC)))

num_intervalos2 = int(1 + 3.322 * np.log10(len(datasetCC)))

print(num_intervalos)
print(num_intervalos2)
```

```python
datasetCC['Intervalos'] = pd.cut(datasetCC['Peso_Estudiantes'], bins=num_intervalos)
datasetCC
```

```python
frecuenciaCC = datasetCC['Intervalos'].value_counts().sort_index().reset_index()
frecuenciaCC
```

```python
frecuenciaCC.columns = ['Peso_estudiantes', 'Frecuencia_absoluta']
frecuenciaCC
```

```python
frecuenciaCC['Frecuencia_acumulada'] = frecuenciaCC['Frecuencia_absoluta'].cumsum()

suma_acumulada = frecuenciaCC['Frecuencia_absoluta'].sum()

frecuenciaCC['Frecuencia_relativa'] = frecuenciaCC['Frecuencia_absoluta'] / suma_acumulada

frecuenciaCC['Frecuencia_relativa_acumulada'] = frecuenciaCC['Frecuencia_relativa'].cumsum()

frecuenciaCC
```

```python

plt.figure(figsize=(10, 6))


plt.plot(frecuenciaCC.index, frecuenciaCC['Frecuencia_absoluta'], marker='o', label='Frecuencia Absoluta')

plt.plot(frecuenciaCC.index, frecuenciaCC['Frecuencia_acumulada'], marker='o', label='Frecuencia Acumulada')

plt.title('Grafica de lineas - Absoluta y acumulada')
plt.xlabel('Índice')
plt.ylabel('Frecuencia')
plt.legend()

plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(frecuenciaCC['Peso_estudiantes'].astype(str), frecuenciaCC['Frecuencia_relativa'], color='skyblue')
plt.xlabel('Peso de los Estudiantes')
plt.ylabel('Frecuencia Relativa')
plt.title('Gráfico de Barras - Frecuencia Relativa')
plt.xticks(rotation=45) 
plt.tight_layout()
plt.show()
```
