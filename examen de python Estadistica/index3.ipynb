```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

la tabla que genera es una tabla que muestra la informacion de la Freuencia absolutas, relativas y acumuladas de la tabla Calificaciones

puede ver mas inofracion en etse video: https://www.youtube.com/watch?v=6d9zYF53Nc8&pp=ygUrRnJldWVuY2lhIGFic29sdXRhcywgcmVsYXRpdmFzIHkgYWN1bXVsYWRhcw%3D%3D

```python

df = pd.read_excel('variables_estadisticas_updated.xlsx', sheet_name='Calificaciones')

data = df.values.flatten()

data_df = pd.DataFrame(data, columns=['Calificación'])

frecuencia_absoluta = data_df['Calificación'].value_counts().sort_index()

frecuencia_relativa = frecuencia_absoluta / len(data)

frecuencia_acumulada = frecuencia_absoluta.cumsum()


resultados = pd.DataFrame({
    'Calificación': frecuencia_absoluta.index,
    'Frecuencia Absoluta': frecuencia_absoluta.values,
    'Frecuencia Relativa': frecuencia_relativa.values,
    'Frecuencia Acumulada': frecuencia_acumulada.values
})

print("Tabla de frecuencia:")
resultados

```

la grafica que genera esto es mostrando los datos Absolutos de la tabla Calificaciones

 puede ver mas información en etse video:    https://www.youtube.com/watch?v=TCrrzZf1GgU&pp=ygUbRnJldWVuY2lhIGFic29sdXRhcyBncmFmaWNv

```python

abs_values = df.apply(pd.Series.value_counts).fillna(0)
abs_values.plot(kind='bar', stacked=True)
plt.title('Valores absolutos de la tabla por tipo de variable')
plt.xlabel('Tipo de variable')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.legend(title='Valor')
plt.show()
```

la grafica que genera esto es mostrando los datos relativos de la tabla Calificaciones

 puede ver mas información en etse video:   https://www.youtube.com/watch?v=PPOEmEVUmdc&pp=ygUaRnJldWVuY2lhIHJlbGF2aXRvIGdyYWZpY28%3D

```python

counts = df.apply(pd.Series.value_counts).T


relative_values = counts.div(counts.sum(axis=1), axis=0)

relative_values.plot(kind='bar', stacked=True)
plt.title('Valores relativos de la tabla por tipo de variable')
plt.xlabel('Tipo de variable')
plt.ylabel('Valor relativo')
plt.legend(title='Valor')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

esta muestra la media, mediana, mínimo, máximo y moda. determinada en la tabla Calificaciones
