```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

la tabla que genera es una tabla que muestra la informacion de la Freuencia absolutas, relativas y acumuladas de la tabla Numero_Estudiantes

puede ver mas inofracion en etse video: https://www.youtube.com/watch?v=6d9zYF53Nc8&pp=ygUrRnJldWVuY2lhIGFic29sdXRhcywgcmVsYXRpdmFzIHkgYWN1bXVsYWRhcw%3D%3D

```python

file_path = 'variables_estadisticas_updated.xlsx'
sheet_name = 'Numero_Estudiantes'
data = pd.read_excel(file_path, sheet_name=sheet_name, header=None)
inicio_tabla = 0
for i, row in data.iterrows():
    if isinstance(row[0], int): 
        inicio_tabla = i
        break
tabla_datos = data.iloc[inicio_tabla:, :].dropna().astype(int)
numero_estudiantes = tabla_datos[0]
frecuencia_absoluta = numero_estudiantes.value_counts().sort_index()
frecuencia_relativa = frecuencia_absoluta / frecuencia_absoluta.sum()
frecuencia_acumulada = frecuencia_absoluta.cumsum()
frecuencia_relativa_acumulada = frecuencia_relativa.cumsum()
tabla_frecuencias = pd.DataFrame({
    'Frecuencia Absoluta': frecuencia_absoluta,
    'Frecuencia Relativa': frecuencia_relativa,
    'Frecuencia Acumulada': frecuencia_acumulada,
    'Frecuencia Relativa Acumulada': frecuencia_relativa_acumulada
})
tabla_frecuencias
```

la grafica que genera esto es mostrando los datos Absolutos de la tabla Numero_Estudiantes

 puede ver mas informaci√≥n en etse video:    https://www.youtube.com/watch?v=TCrrzZf1GgU&pp=ygUbRnJldWVuY2lhIGFic29sdXRhcyBncmFmaWNv

```python

df = pd.read_excel("variables_estadisticas_updated.xlsx", sheet_name="Numero_Estudiantes")
plt.figure(figsize=(10, 6))
for column in df.columns:
    plt.plot(df[column].abs(), label=column)
plt.title('Valores Absolutos de Tipos de Variables')
plt.xlabel('N√∫mero de Estudiante')
plt.ylabel('Valor Absoluto')
plt.legend()
plt.grid(True)
plt.show()
```

la grafica que genera esto es mostrando los datos relativos de la tabla Numero_Estudiantes

 puede ver mas informaci√≥n en etse video:   https://www.youtube.com/watch?v=PPOEmEVUmdc&pp=ygUaRnJldWVuY2lhIHJlbGF2aXRvIGdyYWZpY28%3D

```python


counts = df.apply(pd.Series.value_counts).T


relative_values = counts.div(counts.sum(axis=1), axis=0)

relative_values.plot(kind='bar', stacked=True)
plt.title('Valores relativos de la tabla por tipo de variable')
plt.xlabel('Tipo de variable')
plt.ylabel('Valor relativo')
plt.legend(title='Valor')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

etsa muestra la media, mediana, m√≠nimo, m√°ximo y moda. determinada en la tabla Numero_Estudiantes

mas informacion en este video: https://www.youtube.com/watch?v=RwI159ZBh38&pp=ygUpbWVkaWEsIG1lZGlhbmEsIG3DrW5pbW8sIG3DoXhpbW8geSBtb2RhLiA%3D


```python
columnas_cuantitativas = df.columns
estadisticas = {}
for columna in columnas_cuantitativas:
    estadisticas[columna] = {
        "Media": df[columna].mean(),
        "Mediana": df[columna].median(),
        "M√≠nimo": df[columna].min(),
        "M√°ximo": df[columna].max(),
        "Moda": df[columna].mode()[0] 
    }
for columna, valores in estadisticas.items():
    print(f"Estad√≠sticas para la variable '{columna}':")
    for estadistica, valor in valores.items():
        print(f"{estadistica}: {valor}")
```
