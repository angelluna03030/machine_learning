```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

la tabla que genera es una tabla que muestra la informacion de la Freuencia absolutas, relativas y acumuladas de la tabla Estaturas

puede ver mas inofracion en etse video: https://www.youtube.com/watch?v=6d9zYF53Nc8&pp=ygUrRnJldWVuY2lhIGFic29sdXRhcywgcmVsYXRpdmFzIHkgYWN1bXVsYWRhcw%3D%3D

```python


df = pd.read_excel('variables_estadisticas_updated.xlsx', sheet_name='Estaturas')
data = df.values.flatten()
data_df = pd.DataFrame(data, columns=['Estatura'])
frecuencia_absoluta = data_df['Estatura'].value_counts().sort_index()
frecuencia_relativa = frecuencia_absoluta / len(data)
frecuencia_acumulada = frecuencia_absoluta.cumsum()
resultados = pd.DataFrame({
    'Estatura': frecuencia_absoluta.index,
    'Frecuencia Absoluta': frecuencia_absoluta.values,
    'Frecuencia Relativa': frecuencia_relativa.values,
    'Frecuencia Acumulada': frecuencia_acumulada.values
})
print("Tabla de frecuencia:")
resultados

```

la grafica que genera esto es mostrando los datos Absolutos de la tabla Estaturas

 puede ver mas informaci√≥n en etse video:    https://www.youtube.com/watch?v=TCrrzZf1GgU&pp=ygUbRnJldWVuY2lhIGFic29sdXRhcyBncmFmaWNv

```python
plt.figure(figsize=(10, 6))
for column in df.columns:
    plt.plot(df[column].abs(), label=column)
plt.title('Valores Absolutos de Tipos de Variables')
plt.xlabel('N√∫mero de Estudiante')
plt.ylabel('Valor Absoluto')
plt.legend()
plt.grid(True)
plt.show()
```

la grafica que genera esto es mostrando los datos relativos de la tabla Estaturas

 puede ver mas informaci√≥n en etse video:   https://www.youtube.com/watch?v=PPOEmEVUmdc&pp=ygUaRnJldWVuY2lhIHJlbGF2aXRvIGdyYWZpY28%3D

```python
counts = df.apply(pd.Series.value_counts).T
relative_values = counts.div(counts.sum(axis=1), axis=0)
relative_values.plot(kind='bar', stacked=True)
plt.title('Valores relativos de la tabla por tipo de variable')
plt.xlabel('Tipo de variable')
plt.ylabel('Valor relativo')
plt.legend(title='Valor')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

esta muestra la media, mediana, m√≠nimo, m√°ximo y moda. determinada en la tabla Estaturas

mas informacion en este video: https://www.youtube.com/watch?v=RwI159ZBh38&pp=ygUpbWVkaWEsIG1lZGlhbmEsIG3DrW5pbW8sIG3DoXhpbW8geSBtb2RhLiA%3D

```python
columnas_cuantitativas = df.columns
estadisticas = {}
for columna in columnas_cuantitativas:
    estadisticas[columna] = {
        "Media": df[columna].mean(),
        "Mediana": df[columna].median(),
        "M√≠nimo": df[columna].min(),
        "M√°ximo": df[columna].max(),
        "Moda": df[columna].mode()[0] 
    }
for columna, valores in estadisticas.items():
    print(f"Estad√≠sticas para la variable '{columna}':")
    for estadistica, valor in valores.items():
        print(f"{estadistica}: {valor}")
```
