```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
plt.title("TipoVenta vs Producto")
plt.xlabel("TipoVenta")
plt.ylabel("Producto")
ax.bar(data["TipoVenta"], data["Producto"])
plt.show()

```

```python

fig, ax = plt.subplots()
x = np.linspace(-3.0, 3.0, 100)
y = np.linspace(-3.0, 3.0, 100)
x, y = np.meshgrid(x, y)
z = np.sqrt(x**2 + 2*y**2)
ax.contourf(x, y, z)
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
plt.title("Genero vs Cantidad")
plt.xlabel("Cantidad")
plt.ylabel("Genero")
ax.scatter(data["Genero"], data["Cantidad"])
plt.show()
```
